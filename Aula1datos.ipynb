Desafío

1 - Promedio de área de todos los inmuebles en los barrios en el dataset. El top 10.

2 - Consultar otros datos estadísticos, conteo, mediana, valores mínimo y máximo.

```python
#Promedio de Area de cada Barrio

inmuebles.groupby('Barrio').mean()
```

```python
# Grafico Desafio 1

Prom_barrio = inmuebles.groupby('Barrio')['Area'].mean()

Prom_barrio.sort_values(ascending=False).head(10).plot.barh()

Prom_top_diez = Prom_barrio.sort_values(ascending=False).head(10).plot.barh()

Prom_top_diez
```

```python
#Grafico Pastel

import matplotlib.pyplot as plt

Prom_barrio.sort_values(ascending=False).head(10)

Top_diez = Prom_barrio.nlargest(10)
plt.pie(Top_diez, autopct="%0.1f %%")
plt.axis("equal")
plt.show()
```
